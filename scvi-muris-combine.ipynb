```python
import os
from loompy import loompy
import scanpy.api as sc
import pandas as pd
```

```python
base_10x = "data/tm_10x_v3/droplet"
```

```python
tm_10x = None
df_10x = pd.read_csv("data/tm_10x_v3/annotations_droplet.csv")
df_10x["barcode"] = df_10x.cell.str.split("_").str[-1] + "-1"
sample = ['_'.join(x.split('_')[0:3]) for x in df_10x.cell]
df_10x["sample"] = sample
df_10x["platform"] = "10x"

for tenxdir in os.listdir(base_10x):
    print(tenxdir)
    sample_prefix = tenxdir.split('-')[1]    
    df_sample = df_10x[df_10x["sample"] == sample_prefix].set_index("barcode")
    cols = df_sample.columns

    if tm_10x is None:
        tm_10x = sc.read_10x_mtx(os.path.join(base_10x, tenxdir))
        sc.pp.filter_cells(tm_10x, min_counts=1000)
        sc.pp.filter_cells(tm_10x, min_genes=500)
        for col in cols:
            tm_10x.obs[col] = df_sample[col]
        print(tm_10x)
    else:
        temp = sc.read_10x_mtx(os.path.join(base_10x, tenxdir))
        sc.pp.filter_cells(temp, min_counts=1000)
        sc.pp.filter_cells(temp, min_genes=500)
        for col in cols:
            temp.obs[col] = df_sample[col]
        print(temp)
        tm_10x = tm_10x.concatenate(temp)
```

```python
tm_10x.write_loom("data/tm_10x.loom")
```

```python
base_smart = "data/tm_smart_v8/FACS"
```

```python
tm_smart = None
df_smart = pd.read_csv("data/tm_smart_v8/annotations_facs.csv").set_index("cell")
df_smart["platform"] = "Smart-seq2"
cols = df_smart.columns
for smartfile in os.listdir(base_smart):
    print(smartfile)
    if tm_smart is None:
        tm_smart = sc.read_csv(os.path.join(base_smart, smartfile), first_column_names=True).transpose()
        for col in cols:
            tm_smart.obs[col] = df_smart[col]
        sc.pp.filter_cells(tm_smart, min_counts=50000)
        sc.pp.filter_cells(tm_smart, min_genes=500)
        print(tm_smart)
    else:
        temp = sc.read_csv(os.path.join(base_smart, smartfile), first_column_names=True).transpose()
        for col in cols:
            temp.obs[col] = df_smart[col]
        sc.pp.filter_cells(temp, min_counts=50000)
        sc.pp.filter_cells(temp, min_genes=500)
        print(temp)
        tm_smart = tm_smart.concatenate(temp)
```

```python
tm_smart.write_loom("data/tm_smart.loom")
```
